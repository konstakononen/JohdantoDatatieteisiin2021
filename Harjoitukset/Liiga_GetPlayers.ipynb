```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import bs4
import requests

from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = 'https://www.liiga.fi/fi/tilastot/2020-2021/runkosarja/pelaajat/'
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
```

```python
# Get title 
title = soup.find('title')
print("Avattu: " + title.string)
```

```python


for all_data in soup.find_all('table', {'id': 'stats'}):
    #Otsikkorivi
    for headers in all_data.find_all('thead'):
        for header in headers.find_all('tr'):
            column_names = []
            column_names.append("Kausi")
            column_names.append("Linkki")
            for column in header.find_all('td'):
                column_names.append(' '.join(column.text.split()))
            
            print(column_names)
            
    
    #Pelaajatilastot
    for table_of_stats in all_data.find_all('tbody'):
        list_of_players = []
        for stats in table_of_stats.find_all('tr'):
            
            player = []
            player.append('2020-2021')
            
            link = ""
            for a in stats.find_all('a', href=True):
                link = a.get('href')
            player.append(link)
            
            for detail in stats.find_all('td'):
                player.append(' '.join(detail.text.split()))
                print(player)
                
            list_of_players.append(player)

            
            
            
```

```python
df = pd.DataFrame(list_of_players, columns = column_names) 
print(df)
```

```python
test_list = df['Linkki'].unique()
for test in test_list:
    print(test)
```
