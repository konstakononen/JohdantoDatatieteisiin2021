```python
# Ladataan datasetit
import pandas as pd

df_Players = pd.read_csv('PlayersStats_by_Season.csv')
df_Plaer_Info = pd.read_csv('Player_Information.csv')
df_Players_Salaries = pd.read_csv('PlayersSalaries_by_Season.csv')
```

```python
# Yhdistetään datasetit left joinilla
dataset = df_Players.merge(df_Plaer_Info, on='pelaaja_id', how='left')
print("Taulut yhdistetty ja uusi dataset luotu. \n")
```

```python
# Tarkistetaan datasetti
print("Uusi datasetti: \n")
print(dataset)
print("")
```

```python
# Muokataan tietotyyppejä
dataset['Pelaaja_Kansalaisuus'] = dataset['Pelaaja_Kansalaisuus'].astype('category')
dataset['Peliaika_minuutit'] = dataset['Peliaika_minuutit'].replace(['-'], None)
dataset['Peliaika_minuutit'] = dataset['Peliaika_minuutit'].astype('int64')


print("Tietotyypit: \n")
print(dataset.dtypes)
print("")
```

```python
# Tallennetaan tässä vaiheessa lopullinen setti talteen 
dataset.to_csv(r'C:\Users\35840\Documents\JohdantoDatatieteisiin2021\JohdantoDatatieteisiin2021\Harjoitustyö\Liiga_Players_AllData.csv', index = False, header=True)

print("Tallennettu tässä vaiheessa lopullinen datasetti Liiga_Players_AllData.csv \n")
```
