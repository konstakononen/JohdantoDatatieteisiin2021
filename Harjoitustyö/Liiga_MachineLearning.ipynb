```python
# Ladataan datasetit
import pandas as pd

dataset = pd.read_csv('Liiga_Players_AllData.csv')
```

```python
dataset.dropna(subset = ["Pelaaja_Ikä"], inplace=True) 
dataset.groupby(['Ulkomaille', 'Kausi']).count() 
dataset.groupby(['Pelaaja_Kansalaisuus', 'Kausi']).count() 
```

```python
dataset.corr()
```

```python
print("Luodaan datasetistä ML_df, jota voidaan muokkailla tulevissa kohdissa. \n")

ML_df = dataset 
print(ML_df.dtypes)
```

```python
ML_df.corr()
```

```python
ML_df.corr()['P']

# negatiivinen korreloi käänteisesti, eli pienenee --> tulos suurenee ja päinvastoin
```

```python
ML_df.corr()['Pelaaja_Ikä']
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpls

heatmap_df = ML_df
heatmap_df.drop('Kausi_aloitus', inplace=True, axis=1)
heatmap_df.drop('Kausi_lopetus', inplace=True, axis=1)
heatmap_df.drop('pelaaja_id', inplace=True, axis=1)
heatmap_df.drop('Pelaaja_Pituus', inplace=True, axis=1)
heatmap_df.drop('Pelaaja_Paino', inplace=True, axis=1)
heatmap_df.drop('Plussat', inplace=True, axis=1)
heatmap_df.drop('Miinukset', inplace=True, axis=1)
heatmap_df.drop('PlusMiinus', inplace=True, axis=1)
heatmap_df.drop('AV', inplace=True, axis=1)
heatmap_df.drop('YV', inplace=True, axis=1)
heatmap_df.drop('VM', inplace=True, axis=1)
heatmap_df.drop('A', inplace=True, axis=1)
```

```python
sns.heatmap(heatmap_df.corr(), annot=True, cmap = 'YlGnBu')
plt.savefig('heatmap.png')
plt.show()
```

```python
from sklearn import linear_model

# lineaarinen regressio 

lr = linear_model.LinearRegression()
x = ML_df['P'].values[:,np.newaxis]
y = ML_df['M']

classifier = lr.fit(x,y)

plt.scatter(x,y, color='Green')
plt.plot(x, classifier.predict(x), color='Yellow')
plt.xlabel('Pisteet')
plt.ylabel('Maalit')
plt.title('Pisteiden määrän vaikutus maaleihin')
```

```python
# lineaarinen regressio 

lr = linear_model.LinearRegression()
x = ML_df['P'].values[:,np.newaxis]
y = ML_df['Tulokas']

classifier = lr.fit(x,y)

plt.scatter(x,y, color='Green')
plt.plot(x, classifier.predict(x), color='Yellow')
plt.xlabel('Pisteet')
plt.ylabel('Tulokas')
plt.title('Tulokkuuden vaikutus pisteisiin')
```

```python
# lineaarinen regressio 

lr = linear_model.LinearRegression()
x = ML_df['Tulokas'].values[:,np.newaxis]
y = ML_df['Pelaaja_Ikä']

classifier = lr.fit(x,y)

plt.scatter(x,y, color='Green')
plt.plot(x, classifier.predict(x), color='Yellow')

plt.xlabel('Tulokas')
plt.ylabel('Pelaaja_Ikä')
plt.title('Pelaajan iän yhteys tulokkuuteen')
```
